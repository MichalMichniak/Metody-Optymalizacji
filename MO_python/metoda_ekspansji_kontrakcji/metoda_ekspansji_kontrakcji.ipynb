```python
from metoda_ekspansji import *
import numpy as np
import matplotlib.pyplot as plt
f1 = lambda x:0.01*np.sin(10*x)+(0.07*x**2-0.4)**2
```

```python
import metoda_kontrakcji
x0,x1,n = ekspansion_method(1, 0.5, f1, 1.1 ,10000)
print(x0,x1,n)
x_0,x_1,m = metoda_kontrakcji.contraction_method(x1,x0,f1,0.99,1000)
print(x_0,x_1,m)
plt.plot(np.linspace(-20,20,10000),f1(np.linspace(-20,20,10000)))
plt.plot([x0,x1],[0.025,0.025],'*r')
plt.plot([x_0,x_1],[0.025,0.025],'*b')
plt.xlim([x0,x1])
plt.ylim([0.02,0.03])
plt.show()
```
