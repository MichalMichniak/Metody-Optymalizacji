```python
from metoda_ekspansji import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1 = lambda x:0.01*np.sin(10*x)+(0.07*x**2-0.4)**2

x0,x1,n = ekspansion_method(-8, 2, f1, 1.1 ,10000)
print(x0,x1,n)
plt.plot(np.linspace(-20,20,10000),f1(np.linspace(-20,20,10000)))
plt.plot([x0,x1],[0,0],'*r')
plt.xlim([-4,-1.723143246557993])
plt.ylim([-0.02,0.1])
plt.show()
```
